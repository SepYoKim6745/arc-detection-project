```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV   
file_path = 'one-class SVM/log/realtime_log_ocsvm1.csv'  #    

# CSV  
column_names = ['timestamp', 'v_raw', 'c_raw', 'voltage', 'current']
df = pd.read_csv(file_path, header=None, names=column_names)

#   
plt.figure(figsize=(10, 4))
plt.plot(df['voltage'], label='Voltage')
plt.title('Voltage Waveform')
plt.xlabel('Sample Index')
plt.ylabel('Voltage')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

#   
plt.figure(figsize=(10, 4))
plt.plot(df['current'], label='Current')
plt.title('Current Waveform')
plt.xlabel('Sample Index')
plt.ylabel('Current')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
