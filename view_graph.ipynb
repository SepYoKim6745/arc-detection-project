```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 경로 지정
file_path = 'one-class SVM/log/realtime_log_ocsvm1.csv'  # 파일 경로에 맞게 수정

# CSV 파일 읽기
column_names = ['timestamp', 'v_raw', 'c_raw', 'voltage', 'current']
df = pd.read_csv(file_path, header=None, names=column_names)

# 전압 파형 시각화
plt.figure(figsize=(10, 4))
plt.plot(df['voltage'], label='Voltage')
plt.title('Voltage Waveform')
plt.xlabel('Sample Index')
plt.ylabel('Voltage')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 전류 파형 시각화
plt.figure(figsize=(10, 4))
plt.plot(df['current'], label='Current')
plt.title('Current Waveform')
plt.xlabel('Sample Index')
plt.ylabel('Current')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
